```python
#Plotting Bar Chart
def bar_plot(data_column, x_label, y_label, title, size=(10,8), rotate=0):
    a = data_column.value_counts()
    severity = a.index
    count = a.values
    plt.figure(figsize=size)
    ax = sns.barplot(x = severity, y=count)
    ax.set_xlabel(x_label, fontsize=15)
    ax.set_ylabel(y_label, fontsize=15)
    ax.set_title(title, fontsize=22)
    plt.xticks(rotation=rotate)
    plt.show()
```

```python
#Bar Chart which type of collision & type
bar_plot(df['SEVERITYDESC'], 'Severity', 'Number of Collisions', 'Collision Severity')

#Conclusion by looking at this graph
#You are able to see the difference between the injuries and the property and are able to see that the property damages 
#are much higher wich means that the accidents may not be as severe as those that are propety
```

```python
#Bar Chart that shows the collision & where it happened
bar_plot(df['ADDRTYPE'], 'Address Type', 'Number of Collisions', 'Location of Accidents')
#By looking at this graph, you are able to see that the number of accidents at blocks are much higher than intersection
#This means that more signs or cameras could be implemented so that you could keep an eye out and also so that individuls
#wouldbe more aware that this is an area where accidents are likely to occur
```

```python
bar_plot(df['COLLISIONTYPE'], 'Collision Type','Number of Collisions', 'Collision Type Accidents Chart')
#Here you are able to see that accidents happen a lot more when cars are parked, in order to prevent this you may need
#to implement a parking system that may not be as clsuterred and have a bigger amount of space.
```
